## Calculating Interest Earned

Let's calculate interest payments and FV with python script.

$FV = PV(1+r)^t$

$r = (\frac{FV}{PV})^{1/t}-1$

You can python like a calculator. And conveniently, you can define variables before using them in an equation. This is the strategy that we take in solving for the above equation.

Suppose that two years after you place 1000 dollars in a deposit account, the value o the deposit grows to 1200 dollars. Build python script that calculates the annual interest rate.

```python
FV = 1200
PV = 1000
t = 2
r = (FV / PV) ** (1 / t) - 1
r
```

Let's test this result to make sure that $PV(1+r)^t$ yields the correct FV.

```python
FV_test = PV * (1 + r) ** t
FV_test
```

Interest can be compounded annually (once per year) or it can be compounded more than once. If the interest is compounded more than once annually,  in the present equation we divide the interest by the number times compounded, $n$, and we multiply the number of years, $t$, by $n$.

$FV = PV(1+\frac{r}{n})^{tn}$

Next, create a dictionary that records the value of $1000 after a year of collecting interest over different rates of compounding.

```python
# use dictionary to define how many times interest is compounded
compounding_def = {"annual":1,
                  "semi-annual":2,
                  "quarterly":4,
                  "monthly":12,
                  "weekly":52,
                  "daily":365}

PV = 1000
r = 0.1
t = 1
#compound interest at rate of 10%
# 10% == 0.1
# create dictionary to hold the results
compounding_interest = {}

# for var1, var2 in dict.items()
# for key1, obj1 in dict.items()
for period_name, n in compounding_def.items():
    # use round(val, #decimals) to round value to a specified # of decimals
    compounding_interest[period_name] = round(PV * (1 + r / n) ** (t * n), 2)
compounding_interest
```

Our goal is to teach the computer to search for a solution to this algebra problem.

$ax = 1000$

$a = 3$

Solve for x

1. We will count x upward from 1 until ax > 1000
2. Then we will tell the computer to count in the other direction using smaller increments
     count in negative direction until ax < 1000
3. Then count upward using a small increment . . . 

```python
# solve ax = 1000
a = 3
# we don't know x, so we will count x upward
x = 1
ax = a * x

# We know that we want to count x upward until ax crosses 1000,
# then we want to count downward at lower rate
# i.e., at first count upward by 1, then count downward by 0.1, 
#       then upward again by 0.001, etc...
margin = 1.0

# create variables to keep track of ax's value
# relative to the target solution of 1000
below_1000_currently = True
# we also need to know whether ax was above or below
# 1000 last period, to know whether or not the direction
# of the change in x needs to switch
below_1000_last_period = True

# if ax is approximately 1000, break out of while loop
while round(ax, 10) != 1000:
    
    # if ax < 1000, count downwrad
    # if ax > 1000, count downward
    x = x + margin if ax < 1000 else x - margin
    # after changing the value x, update the value of ax
    ax = a * x
    
    # update whether ax is above or below 1000
    below_1000_currently = True if ax < 1000 else False
    
    # if ax crossed from above 1000 to below 1000
    # or if ax crossed from below 1000 to above 1000
    # shrink the margin by which change x
    if below_1000_currently == True and below_1000_last_period == False:
        margin = margin * 0.1
        print("Start counting upward at rate of", margin)
    if below_1000_currently == False and below_1000_last_period == True:
        margin = margin * 0.1
        print("Start counting downward at rate of", margin)
    
    # once we have identified if ax crossed 1000,
    # update last periods position to match current
    below_1000_last_period = below_1000_currently
    print("x:", x,"ax:", ax)
print("You solved ax = 1000, a = 3, for x")
print("x =", round(x, 2), "ax =", round(ax, 2))
```

```python
a = 3
x = 1
ax = a * x
# solve for ax = 1000
margin = 1.0


# create variable to track if ax is ablove or below 1000
below_1000 = True

# if below_1000 switches from True to False or from False to True,
# indicate that the program needs to change direct 
# (from negative to positive or from positive to negative)
decrease_margin = False

while round(ax,10) != 1000:
    if decrease_margin:
        margin = margin * .1
        decrease_margin = False
    # increase x or decrease x depending on whether
    # x is below or above 1000
    x = x + margin if ax < 1000 else x - margin
    ax = a * x
    print("x =", x, "\tax", ax)
    # if ax crosses over 1000, identify
    # 1) that ax is now greater than 1000
    # 2) x should change by a smaller margin and in a diff
    # direction
    if below_1000 == True and ax > 1000:
        #change by smaller margin
        decrease_margin = True
        # move in opposite direction
        below_1000 = False
    elif below_1000 == False and ax < 1000:
        # change by smaller margin
        decrease_margin == True
        # move in opposite direction
        below_1000 = True
    
print("Solution: (" + str(a)+")" + "("+str(x)+") =", ax)
```

### Use brute force to solve for the interest rate you would need to grow 1000 dollars to 1500 dollars after 2 years of annual compounding

```python
# set up equation
PV = 1000
FV = 1000
# checking interest rates from 0.01 and above 
# 0.01 is 1%
r = 0.01
# margin should increase or decrease by no more than 1%
margin = .01
# PV = FV / (1 + r/n)**(t*n)
# FV = PV (1 + r/n) ** (t/n)

# identify whether FV above or below 1500
below_1500 = True

# set decrease_margin False
# decrease_margin will only be True if either: 
#            below_1500 == True and FV > 1500
#            or
#            below_1500 == False and FV < 1500
# Once decrease margin is true, margin shrinks
# and decrease_margin is again set to False
decrease_margin = False

while round(FV, 2) != 1500:
    if decrease_margin:
        margin = margin * 0.1
        decrease_margin = False
    # r increases by margin if below 1500
    # r decreases by margin if above 1500
    r = r + margin if FV < 1500 else r - margin
    # recalculate future value given change in r
    FV = PV * ( 1 + r) ** 2

    print("r:", r, "\tFV:", FV)
    if below_1500 == True and FV > 1500:
        decrease_margin = True
        below_1500 = False
    if below_1500 == False and FV < 1500:
        decrease_margin = True
        below_1500 = True
print("r:", r, "FV:", round(FV, 2))
```

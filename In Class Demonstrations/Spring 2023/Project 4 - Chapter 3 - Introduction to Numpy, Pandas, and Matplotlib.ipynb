```python
# Create articial data

macro_dict = {"GDP": {},
             "Money": {},
             "Real GDP":{},
             "Price Level":{}}
for key in macro_dict.keys():
    for i in range(1990, 2021):
        macro_dict[key][i] = np.random.random() * 10000
macro_dict
```

```python
macro_DF = pd.DataFrame(macro_dict)
macro_DF
```

```python
macro_DF[["GDP"]]
```

```python
#create a new variable
macro_DF["Velocity"] = macro_DF["Price Level"].mul(macro_DF["Real GDP"].div(macro_DF["Money"]))
macro_DF
```

## matplotlib

```python
import matplotlib.pyplot as plt
macro_DF.plot()
```

```python
# plot using method for matplotlib from the pandas DataFrame
fig, ax = plt.subplots(figsize = (20,12))
macro_DF.plot(linewidth = 5, alpha = .7, legend = False, ax = ax)# color = "C0" . . .
ax.legend(fontsize = 20)
```

```python
# our goal is to create a multiplot visualization with a separate plot area for each variables
# count the number of variables
num_vars = len(macro_DF.keys())

# create a plot with a subplot for each variable (total 5)
fig, axs = plt.subplots(1, num_vars, figsize = (20,6))

# to plot each variable on each plot, cycle through index of the keys
for i in range(num_vars):
    
    ax = axs[i]
    # call key by index number
    key = list(macro_DF.keys())[i]
    # set color in light of number i so that it is "Ci"
    color = "C" + str(i)
    
    alpha = .2 + .15 * i
    # plot key according to index number in ax
    macro_DF[key].plot.line(color = color, alpha = alpha, ax = ax)
    ax.set_title(key,color = color, fontsize = 20)
```

## Plotting Income Distribution

```python
income_data = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls",
                            header=[7])
income_data = income_data.rename(columns = {"Unnamed: 0": "Household Income"})
income_data
```

```python
income_data.iloc[1:-1]
```

```python
# save series of persons in each income level
income = income_data["Number"][1:-1].div(1000)
income.iloc[0] = 0
income.head()
```

```python
# Make a list that will serve as the index to replace current index
income_levels = list(range(0, 201, 5))
income_levels.append(250)
income_levels = [str(val) + "k" for val in income_levels]
income_levels.append("250k+")
income_levels
```

```python
income_data = pd.DataFrame({"Income Level":np.array(income_levels),
                           "Households in Bracket":income})
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()
income_data.set_index("Income Level", inplace = True)
income_data
# income_data.plot.line(secondary_y = "Cumulative")
```

```python
total_households = income_data["Households in Bracket"].sum().round(3)
income_data["% in Bracket"] = income_data["Households in Bracket"].div(
    total_households).mul(100)
income_data["Cumulative %"] = income_data["% in Bracket"].cumsum().round(3)
income_data
```

```python
plt.rcParams.update({"font.size": 16})
plot_df = income_data[["% in Bracket", "Cumulative %"]]
fig, ax = plt.subplots(figsize = (15, 7))
plot_df.plot.bar(secondary_y = "Cumulative %",
                 width = .95, 
                 alpha = .75,
                 ax = ax)
x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
```

```python
# build a distribution of income with 25k bins instead of 5k bins
# start by counting by through rows by 5s in the original dataframe 
income_25k_bin = income_data.loc[::5]
# we were missing the final row with 250k+, so add this in the last row
# we created a new row using df.loc[new_row_name]
income_25k_bin.loc["250k+"] = income_data.iloc[-1]
# redefine the marginal bins by taking the difference of the cumulative bins
income_25k_bin["Households in Bracket"] = income_25k_bin[
    "Cumulative"].diff().fillna(0)
# the information from the % in Bracket column was incorrect. So, we 
#  recalculate this using the new Households in Bracket data
income_25k_bin["% in Bracket"] = income_25k_bin["Households in Bracket"].div(
    total_households).mul(100)
income_25k_bin
```

```python
plt.rcParams.update({"font.size" : 20})
plot_df = income_25k_bin[["% in Bracket", "Cumulative %"]].iloc[1:]
fig, ax = plt.subplots(figsize = (15, 10))
plot_df.plot.bar(secondary_y = "Cumulative %",
                 width = .99, 
                 alpha = .75,
                 color = ["C0","C1"],
                 ax = ax)

x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
ax.set_title("Income Distribution in the United States")
```

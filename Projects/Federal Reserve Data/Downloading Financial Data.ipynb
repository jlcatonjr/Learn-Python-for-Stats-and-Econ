```python
import numpy as np
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.backends.backend_pdf import PdfPages

ticker = "^GSPC"
start = datetime.datetime(1975, 1, 1)
# end = datetime.datetime(1985, 12, 31)
end = datetime.datetime.today()
freq = "D"
SnP= web.DataReader(ticker, start = start, end = end, 
                    data_source = "yahoo").resample(freq).mean()
SnP.dropna()
```

```python
plt.rcParams.update({'font.size': 32})
plt.rcParams['axes.ymargin'] = 0
plt.rcParams['axes.xmargin'] = 0
SnP["Close"].plot(figsize = (16,10))
```

```python
tickers = "MSFT", "AAPL", "IBM", "GOOG"
data_dict= {}
start = datetime.datetime(2000, 1, 1)

for stock in tickers:
    data_dict[stock] = web.DataReader(stock, "yahoo", start, end)

data_dict
```

```python
# make a list of the close data for each stock
close_data_dict = {key:val["Close"] for key, val in data_dict.items()}
close_data_df = pd.DataFrame(close_data_dict)
close_data_df
```

```python
#from datlib.plots import *
close_data_df.plot(legend = False)
```

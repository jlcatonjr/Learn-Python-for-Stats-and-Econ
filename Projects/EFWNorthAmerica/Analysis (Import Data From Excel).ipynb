```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data_cats = ["NoTaxRatios","WithTaxRatios"]
dcts = {}
for data_cat in data_cats:
    filename = f"panelData{data_cat}.xlsx"
    dcts[data_cat] = pd.read_excel(filename,sheet_name=None)
    for key, df in dcts[data_cat].items():
        dcts[data_cat][key].set_index(["State","DATE"], inplace = True)
        del dcts[data_cat][key]["Unnamed: 0"]
dcts
```

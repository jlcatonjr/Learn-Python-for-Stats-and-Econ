```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import uniform
a = 2
b = 50
ab = np.linspace(a,b,10**6)
values = 1 / ab[-1::-1]
values = pd.DataFrame(values, columns = ["Distribution"])
```

```python
plt.rcParams.update({'font.size': 26})
plt.rcParams['axes.ymargin'] = 0.0
plt.rcParams['axes.xmargin'] = 0.0

fig, ax = plt.subplots(figsize = (12,6))
values.plot.hist(ax = ax, bins = 10000, legend = False, histtype='stepfilled', density = True)
ax.set_title("Inverse Uniform Distribution")
ax.text(-0.01,-1,"0.02", ha = "left", va = "top")
ax.set_yticklabels([])
ax.tick_params('both', length=0, which='both')
```

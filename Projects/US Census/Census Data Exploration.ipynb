```python
full_search = pd.DataFrame(censusdata.search('acs5', 2015, 'label', '')).set_index(0)
```

```python
codes = full_search[full_search.index.get_level_values(0).str.contains("_001")]
codes
```

```python
codes.to_csv("CensusSearchCodes.csv")
```

```python
download_codes= {ix: codes.loc[ix][1] for ix in codes.index}
pd.DataFrame(download_codes.values(), index = download_codes.keys())
```

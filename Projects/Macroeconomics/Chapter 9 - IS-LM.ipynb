# IS-LM

## Credit Markets and Real Income in the Short-run

So far, we have analyzed the macroeconomic impact of changes in the supply of and demand for money. In the short-run, changes in the supply of or demand for money can impact the value of real income produced in the economy. Using the equation of exchange, we showed that these fluctuations lead to a mismatch between the value of aggregate demand – total expenditures across an economy – and the value of long-run aggregate supply (which in equilibrium is the same as the value of real output available to serve as real income). If the value of aggregate demand is greater than the value of available aggregate supply, entrepreneurs may interpret the increase in expenditures as an increase in demand for their products, thus leading them to provide more real goods to the market. If the value of aggregate demand is less than the value of aggregate supply available for sale, businesses may interpret the discrepancy as a diminishment in demand for their goods, leading them to produce less than they otherwise would. These short-run responses are only offset as prices adjust to reflect changes in aggregate demand factors. Prices tend to rise in the case of a positive aggregate demand shock and fall in the case of a negative aggregate demand shock.

Last chapter, we expanded our definition of the money stock to include financial instruments, recognizing that these instruments provide access to liquidity and, therefore, serve a monetary function. We emphasized the link between the velocity of money and the interest rate, showing that increases in velocity tend to be positively associated with increases in the rate of interest (this is true for both real and nominal interest rates). Similar tendencies are built into another macroeconomic model of short-run phenomena: IS-LM. We will review the early history and motivation of the model, present the model, and relate the model to fiscal and monetary policy.

## John Maynard Keynes and IS-LM

The IS-LM model is an attempt to link financial markets, money supply and demand, and real income. John Hicks developed the model as an attempt to formally model the insights elaborated by John Maynard Keynes in his General Theory of Employment, Interest, and Money. “The postulates of the classical theory,” Keynes wrote, “are applicable to a special case only and not to the general case (3).” In the General Theory, Keynes presents a framework for approaching short-run macroeconomic fluctuations where capital and labor become inefficiently allocated relative to the long-run equilibrium that is predicted by the classical macroeconomic model.

So far, we have seen that the equation of exchange is capable of representing short-run fluctuations in macroeconomic activity. These fluctuations, driven by changes in aggregate demand factors, lead to discrepancies between the level of real income produced and the level that can be sustainably produced, represented by the long-run aggregate supply curve. The equation of exchange, however, does not integrate the interest rate. The interest rate plays a fundamental role in directing resources toward their most productive uses across time. The creation of credit, as we have seen, impacts the velocity of the base money stock. And we have ssen that the velocity of broadest money stock is strongly and postively correlated with changes in the interest rate. That is, the velocity of money reflects expected returns to investment. The critiques of macroeconomics and policy suggestions from John Maynard Keynes focus on financial markets, so they must be included in our analysis.

## Say’s Law

Keynes launched a fundamental critique toward classical Macroeconomics. The most essential part of this was a rejection of Say’s Law. Say's Law states which states that demand can only be manifested and sustained by provision of valuable goods and services to the market. Keynes reformulated this, writing that “supply creates its own demand” (Chapter 2 General Theory). Keynes argued that 1) markets can oversupply goods, leading to a productive glut where goods are produced but not able to be sold. In other words, oversupply creates economic depression. He argues instead 2) that “demand creates its own supply”, meaning that increases in aggregate demand can support higher values of real income. Short-falls in aggregated demand are, according to Keynes, the result of the failure of financial markets to transfer unutilized resources toward productive uses.
What is Say’s Law and why does it matter? In A Treatise on Political Economy, Say argues

A man who applies his labour to the investing of objects with value by the creation of utility of some sort, can not expect such a value to be appreciated and paid for, unless where other men have the means of purchasing it. Now, of what means do these consist? Of other values of other products, likewise the fruits of industry, capital, and land. Which leads us to a conclusion that may at first sight appear paradoxical, namely, that it is production which opens a demand for products.

According to Say's Law, goods must pay for goods. If someone wishes to purchase a product, he or she must either exchange another good directly for the desired good or else acquire by exchange money to purchase the item. A good offered in exchange for another must be positively valued by the one receiving it. If not, no exchange will occur. Thus, it is not simply that one supplies goods to the market in order to satiate demand for other goods, but that one must supply goods valued by others. Supply of value to the market allows the producer to procure a share of the value provided.

Confusion arises when money must be integrated into the framework. Money, unlike other goods, comprises one side of every exchange in a monetary economy. Since money is integral to all exchanges, fluctuations in its demand and supply can also effect the ability of others to acquire the money requisite to exchange.

Unfortunately, Say does not do a good job of explaining the significance of money within this schema.

Thus, to say that sales are dull owing to the scarcity of money, is to mistake the means for the cause; an error that proceeds from the circumstance, that almost all produce is in the first instance exchanged for money, before it is ultimately converted into other produce: and the commodity, which recurs so repeatedly in use, appears to vulgar apprehensions the most important of commodities, and the end object of all transactions, whereas it is only the medium. Sales cannot be said to be dull because money is scarce, but because other products are so. Should the increase of traffic require more money to facilitate it, the want is easily supplied, and is a strong indication of prosperity – a proof that a great abundance off values has been created, which it is wished to exchange for other values.

Say’s description implies that he understands that there can be an excess demand for money that raises money's price. That increase in price will encourage an increase in the production of money under a commodity standard, but this process takes time. Under the gold standard, for example, production of gold adjusted to economic conditions with a lag of about 2 years. Having brushed off the problem of insufficient aggregate demand by relying on an invisible-hand process, Say argues that dearness of some commodities is often due to overproduction of other commodities. Thus, there can be no general glut, only gluts in particular markets offset by relative scarcity in other markets.

It is because the production of some commodities has declined, that other commodities are superabundant. To use a more hackneyed phrase, people have bought less, because they have made less profit; and they made less profit for one or two causes; either they have found difficulties in the employment of their productive means, or these means have themselves been deficient.

On this account say is correct, however, a glut of money or its dearness could lead to offsetting scarcities and gluts, respectively, in other markets. Say describes a misallocation of resources that adjustment of prices, and subsequently, the capital structure can fix.

### A Reinterpretation of Say’s Law

It is possible however, that the processes that coordinate market activity might be interrupted by extreme swings in demand for money. It is this problem for which Leijonhufvud and Clower’s extension of Say’s Law as “Say’s Principle” (they refer to it as SP) provides a clear explanation. They argue that “the net value of an individual’s planned trades is identically zero”, meaning that you can only plan to spend as much income as you have in your possession. Every week you must make decisions concerning how to allocate your income. Maybe you apportion 20% of your income to savings, 50% to housing and food, leaving the remaining 30% for discretionary expenditures.

In the case that you decide to save money, maybe you hold 5% of your income in the form of cash. Your demand to hold as cash income that you will not spend represents a positive demand to hold money, described in earlier presentations of aggregated demand in terms of portfolio demand.

Algebraically, the authors represent this elementary budget constraint of an individual under a regime of exchange with secure property rights:

>$(\sum_{i=1}^{n} p_{i}q_{i}) + d_m +s_m$

n order for an individual's plans to be coherent, planned expenditures. ($\sum_{i=1}^{n} p_{i}q_{i}$) plus planned holdings of money ($d_m$) must equal the stock of money available to the agent ($s_m$). In order for macroeconomic equilibrium to be possible, the sum of all planned expenditures for all individuals in the economy must perfectly match the nominal value of all available goods. This is only possible if the total value of planned expenditures and demand to hold money over a given period is exactly equal to the total supply of money:

>$\sum_{j=0}^{k} Py_{j} + d_{m, j} - s_{m, j}$

For all individuals, the sum of planned expenditures and planned money holding  must equal exactly the amount of money available to the individual. Trouble arises when the aggregate value of planned expenditures falls short of the value of goods available for sale. Mismatches between these lead to excess demands for money and corresponding excess supplies of goods. If preferences are fixed, this aggregate discrepancy can only be offset by adjustment in prices or changes in the quantity of money.

>Figure 2 Excess Supply of Goods

When money is included in analysis, we find that all excess supplies are equally offset by excess demands for goods. This is of special significance if one is to understand macroeconomic fluctuations. Keynes’s presentation of Say’s Law holds firm to the idea that prices must adjust in order for the economic equilibrium to emerge. Of course, this was a perspective that Say explicitly supported. Because price adjustments take time, fluctuations in aggregate demand, generated by oscillations in financial markets, can lead to displacement of formerly employed labor and capital for a sustained period. Unlike Say, Keynes proposed that this problem should be offset not  by market processes, but by expansionary fiscal policy.  

### Savings = Investment?: Keynes and Say’s Law

Two solutions to this problem have been discussed. Either prices can fall to alleviate growing inventories or the money stock can increase. This has policy implications. A central bank can attempt to alleviate, either in whole or in part, fluctuations in demand for money by increasing the quantity of base money in circulation. A popular formulation of this proposal is that the central bank should attempt to stabilize MV by adjusting M to offset changes in V. Keynes, however, argued that monetary policy was incapable of solving the problem. He believed that a fall in V was, in the extreme case, self-reinforcing in the face of increases in M.

Keynes argued that one should not assume that the macroeconomy will reach its predicted equilibrium state. There may arise a discrepancy between savings and investment leaving some resources are left idle due to hoarding. This occurred primarily in the form of increasing one’s holdings of currency rather than saving in a bank where it would be lent. Further, he believed that even in the case that it was saved in a bank, oscillations in the confidence of economic expectations would typically lead to suboptimal levels of investment coinciding with excessive preference for money-holding. This would prevent the formation of macroeconomic equilibrium predicted by the quantity theory of money as prices, especially wages, may take time to adjust to a negative shock to aggregate demand.

While many macroeconomists would have prescribed increases in the quantity of money to remedy shortfalls in aggregate demand, Keynes believed that this approach was insufficient. The government, he argued, should increase its level of expenditures in order to bring underutilized resources into employment. A government could increase expenditures during economic depression owing to its dependability as a borrower. Aided by expansionary monetary policy aimed at supporting government borrowing, Keynes argued that expansionary fiscal policy could offset depression. While we will not discuss the details of monetary policy until the following chapters, the historical context is useful for understanding the motivation of development of the IS-LM model.

# IS-LM

## Keynesian Cross and the Marginal Propensity to Consume

Keynes used the postulate of a marginal propensity to consume to guide his reader in his opposition to Say’s Law. Keynes motivated his model by description of agent preferences to consume as being linear with regard to income. From this assumption naturally emerges a general glut. Even if consumer incomes rose, hoarding of resources would lead to a gap between output available for sale and total expenditures in cases where production rose above the level that can be sustained by a given value of total expenditures. Thus, the model assumes that productivity generates a glut or, in other words, that an aggregate demand shortfall is a persistent macroeconomic problem.
This is expressed by an aggregate demand curve not strictly defined by $MV$. Keynes simply proposes that aggregate demand is a linear function dependent upon the inclination of actors to consume.

>$AD = \alpha + MPC(y)$

he marginal propensity to consume, MPC, represents the percentage of income that the prototypical agent consumes, with the caveat that when nothing is produced, the agent still borrows in order to consume a value of $\alpha$. Notice that as output for a given period, which is identical to income, rises, expenditures may not keep pace. Thus, the model reflects Keynes’s skepticism that “supply creates its own demand”. For Keynes, this problem does not arise due to a mismatch between consumer demand and he provision of goods by suppliers, as Say would have it, but rather it is due to an inclination toward hoarding. The Keynesian cross presumes that if there is a discrepancy between aggregate demand and long-run aggregate supply, the price level will not adjust to offset this discrepancy.

> Figure 3 Keynesian Cross

Suppose, for example, that the marginal propensity to consume is .5, meaning that half of all income generated is spent and that this matches aggregate demand at 16 trillion dollars. If potential income totals 20 trillion dollars, the model predicts that only 4 trillion dollars worth of GDP will not be realized. According to the model, the economy is stuck in a persistent sub-optimal equilibrium.

Keynes suggested that the discrepancy could be offset by an increase in government expenditures supported by deficits (borrowing). This would, he argued, effectively translate increases in the quantity of money into increases in expenditures, transforming unpurchased output into real income. In other words, Keynes argued that fiscal stimulus would move the economy away from any suboptimal equilibrium. It would do this, he argued, on two margins: 1) the expenditures would themselves add to GDP and 2) they would multiply when new incomes were again spent. This multiplication is known as the Keynesian multiplier.

$KM=\frac{1}{1-MPC}$

If the marginal propensity to consume is .5, as in the example above, then the Keynesian multiplier is 2. An increase in deficit spending of 2 trillion dollars would double to 4 trillion dollars by this logic. GDP increases from 16 trillion dollars to 20 trillion dollars.

### Muliplier Caveat

In order for the multiplier to actually work, the increase in government expenditures, $G$, must not proceed in the form of taxes, which diminishes consumption, $C$, or in the form of borrowing from the market (future taxes) draws down the value of I, investment. Recall that $Y=C+I+G$. In the Solow model, we defined consumption as $Y−I$. With the inclusion of present taxation, we write this as $C=Y-I-T$.

The Keynesian formula works most pristinely when increases in government expenditures are supported by monetary expansion. That is, as new government debt is created, new money is created to support that debt. It is possible that, due to the belief that the state is a dependable borrower, that the crowding out of private investment by government is less than a ratio of 1:1, in which case deficit funded expenditures would move the aggregate demand curve. This would allow fiscal expansion to autonomously exert a positive increase in total expenditures, thus increasing the velocity of the monetary base. These cases will be conveyed in the presentation of the IS-LM model.

### Investment-Savings

Keynes argument relies on the assumption that not all savings is translated into investment. Suppose, with reference to the above presentation, that for one year there was an unexpectedly large level of productivity of 20 trillion dollars and that this level is the true level of productivity that can be sustained within the existing economic arrangement (Figure 4). According to Keynes, individuals would not successfully invest the excess in new forms of production. The extra income (above 16 trillion dollars) would be saved, but it would not be transferred to new productive activities that sustain high levels of income in the following year. The greater productive potential would be wasted due to frugality.

Keynes broke with the neoclassical assumption that $S=I$ (as we assumed with the Solow model). In essence, Keynes assumed 1) that prices do not swiftly equilibrate to promote macroeconomic equilibrium and 2) that individuals have an insatiable demand for holding money, particularly during economic depression. In short, Keynes believed that markets do not promote rational outcomes at the macroeconomic level due to wasteful hoarding - savings not translated into investment. As mentioned above, his second assumption suggested that monetary policy is inadequate for offsetting monetary disequilibria in the form of excess demand for money. Only increases in government expenditures could successfully increase transactions demand for money, thereby lifting productivity and interest rates.

## IS-LM Model

Keynesian theory lives on in the form of the IS-LM model. IS-LM is a model of the short-run that compares the level of real income to the interest rate. It was first presented by Sir John Hicks, using a nominal, as opposed to real, rate of interest. In our presentation, we will use real interest rates, and therefore assume that changes presented in the model are self-reversing upon adjustment of the price level to its long-run level. Thus, we maintain the long-run neutrality of money and assumption of long-run macroeconomic equilibration.

### Investment-Savings

>Figure 6: IS Curve

The IS curve presents investment as a function of the real rate of interest. We can think of investment, in this case, as willingness to borrow. As interest rates fall, investors are willing to increase their level of borrowing intended for productive projects. A shift in investor expectations may move the IS curve. For example, when investors are increasingly pessimistic, the IS curve will shift to the left. When they are increasingly optimistic, the IS curve will shift to the right. An increase in the IS curve represents an increased willingness to engage in project development at a given real rate of interest. During an economic boom, for example, the IS curve tends to shift to the right. During a recession, it tends to shift to the left.

### Liquidity-Preference and Money

Last chapter we related the velocity of money to the interest rate, noting that as the interest rate increases, so too does the average rate of expenditure of each monetary unit. Thus, as the IS curve increases, it moves along the LM curve. An autonomous increase in transactions demand, $Py$, accompanies an increase in the velocity of money as individuals transform money holdings that comprise portfolio demand for money into expenditures and investment which translate to increases in $y$. The LM curve suggests that the transformation of non-invested resources into investment is motivated by an increase in the rate of interest. Thus, the LM curve is upward sloping.

> Figure 7: LM Curve

### Full Model

Together, preferences of individuals to consume and invest, represented by the IS curve, interact with the LM curve which represents the impact of a given quantity of money on income as preferences to hold cash – liquidity preference. These forces correlate interest rates and real income.

>Figure 8:IS-LM Curves

We can think of shifts in the IS curve as being driven by changes in velocity. This is equivalent to velocity driven increases in aggregate demand leading to short-run improvements in income. As velocity increases, so too does the IS curve (Figure 9). Once the price level adjusts, however, economic boom moderates. This is reflected by a return of the IS curve to its original position, presuming that the long-run aggregate supply curve has not also shifted and that there exists no misallocation of resources that induce inefficient arrangements in the existing capital structure.

>Figure 9: Investor Optimism Drives Economic Boom

Likewise, economic recession is driven by a reduction in investment (Figure 10). Perhaps investors expect that the near future will not be especially prosperous. If, for example, many believe that the there has been over investment in one or many sectors, investors will begin to reduce their exposure, perhaps moving to safer investments or holding on to cash. This results in a fall in the velocity of money and, equivalently, a leftward shift in the IS curve.
Although an economic boom may reverse due to tightening liquidity constraints and temporarily high interest rates, these high rates quickly turn low as pessimism concerning returns on investment drive lower levels of investment. This is represented as a leftward shift in the IS curve that includes a fall in rates and a fall in real income.
Note that in either case, the aggregate demand curve shifts in the same direction as the IS curve. The IS curve shifts as individuals become more or less willing to spend and invest their income. The IS curve returns to it’s initial, long-run position as the price level adjusts to offset any excess demands for or excess supplies of goods.

## IS-LM and Fiscal and Monetary Policy

Next we will show that expansionary fiscal policy supported by deficit spending shifts the IS curve to the right. Likewise, expansionary monetary policy will shift the LM curve to the right. Deficit spending may be enabled by monetary expansion, so these two curves often shift together.

>Figure 11: Deficit Spending Increases

In terms of the expenditure approach to GDP, an increase in government spending, $G$, that is not offset by a fall in other variables promotes an increase in the IS curve. Such an expansion in $G$ raises real GDP, $y$, during the period of expenditure (Figure 11).  Assuming that the government finances this through borrowing from private lenders, the increase in borrowing will increase the real rate of interest, $r$. Note that the effects of fiscal expansion only occur during the present period. The IS-LM model is unable to capiture any second-order effects that may arise due to distortion of the capital structure, which entails resource misallocation that may effect the level of real income in future periods.

> Figure 12: Federal Reserve Increases the Quantity of Money

Similarly, central banks, which for all intents and purposes are government entities, may also attempt to temporarily boost real GDP. They typically attempt to accomplish this by an increase in the quantity of base money (though we will consider other means in a later chapter) (Figure 12). The increase in the quantity of money amounts to a rightward shift in the LM curve. This may temporarily increase real GDP and lower the real rate of interest. If investors, however, fully anticipate such an increase, it is possible that the increase in the quantity of money simply raises the price level without impacting y or r. In such a case, long-run effects would immediately swamp any short-run effects.

> Figure 13: Monetary Expansion Supports Deficit Spending

Keynes supported a fiscal expansion supported by money creation from the central bank. This leads to a simultaneous increase in both the IS and LM curves, thereby limiting the effect of government borrowing that would typically raise the real rate of interest (Figure 13).

## Debt-financed expenditures by the U.S. Government

Deficit spending as a macroeconomic policy tool presupposes that without support from the government, recessions may lead to detrimental economic and social outcomes. In particular, unemployment during a deep recession, like in 2008, can impact long-term trends in the labor market as discouraged workers who remain unemployed may fail to return to the labor market.

In recent crises, we have observed governments practicing the Keynesian prescription. The United States experienced recessions in 2001 after the attacks of September 11 and during the financial crisis of 2008 (Figure 4). In both cases, the Federal Government tended to increase expenditures in absolute terms and as a proportion of U.S. GDP. More recently, we have seen the largest episode of deficit spending in American history. The COVID-19 pandemic received a substantial response from congress in the form of a 2.2 trillion dollar spending package, in addition to normal operating expenditures of 4.87 trillion dollars. To put this in perspective, nominal GDP in 2019 was 21.73 trillion dollars. 

One way to measure the significance of fiscal policy is to measure changes in government indebtedness as a percent of nominal GDP (Figure 14). After the 2001 recession, federal debt increased modestly as a proportion of nominal income. The financial crisis of 2008, on the other hand, saw a near doubling of federal indebtedness as a proportion of nominal GDP. The response, in both cases, was in proportion to the significance of the recession.

>Figure 14 Real GDP and U.S. Federal Debt as a \% of GDP

Some economists have challenged any approach that uses fiscal policy as a tool for macrostabilization. They argue that consumers and investors will account for an increase in future taxation that must necessarily support increased indebtedness of the government. This argument is known as Ricardian equivalence. This crowding out may invest resources in projects that yield lower returns than other available projects and, therefore, a lower rate of economic growth. Unlike investors who are subject to market pressures, there is little in the way of mechanisms to correct relatively low-yielding investment by government. 

Another way to approach fiscal policy would be to consider long-term constraints. While fiscal and monetary policy can influence the real rate of interest in the short-run, in the long-run these are determined by market forces. So while fiscal policy might influence nominal income and interest rates in the short-run, in the long-run, public expenditures are constrained by these variables. The higher the interest rate, the more costly it is for government to engage in fiscal expansion. For the last three decades, interest rates have been on the decline, with government spending significantly increasing as rates moved closer to zero. (Figure 15).

>Figure 15 Real Treasury Rates and Real U.S. Debt 

The higher the rate of interest, the greater the burden of debt repayment by the government. Thus, if interest rates begin to increase, it is possible, it will become increasingly difficult for the federal government to maintain it’s current level of debt and annual expenditure. In the least, this could place an upper bound on the level of federal indebtedness compared to GDP or force the government to engage in increasingly activist monetary policy. Even without rising rates, we have seen the latter option and will explore this in the chapter on monetary policy.

# Conclusion

The IS-LM model is yet another means of presenting short-run dynamics. It directly links real income and interest rates, separating changes in demand to spend or hold money (IS) and changes in the quantity of money itself (LM). Movements of the LM curve in particular are self-reversing as the price level adjusts to its long-run value and nominal interest rates adjust to reflect changes in the quantity of money. We have also shown how the IS-LM model relates to the macroeconomic model that we have learned before this chapter. In essence, IS-LM presents another view of the aggregate demand. We have also seen how fiscal policy can impact the level of real income and considered limitations to this approach.

Our presentation of fiscal policy is necessarily incomplete as we have not yet formally introduced monetary policy. Monetary policy often interacts with fiscal policy as the central bank purchases debt from the government with newly created money. It also targets interest rates. In recent years, the Federal Reserve has a developed a variety policies that it can use to support federal borrowing while limiting the inflationary impact of money creation. We will review these in turn.

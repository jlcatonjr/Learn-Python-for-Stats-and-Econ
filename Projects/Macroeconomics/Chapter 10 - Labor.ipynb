## What is Labor?

The macroeconomy is typically studied in terms of different sectors. So far, we have considered how capital markets and the market for money influence other macroeconomics variables. The level of effective capital influences the level of real income per capita that is attainable in the long-run. Money can impact the level of real income in the short-run as fluctuations in demand for and the supply of money lead to short-run imbalances between the amount of money made available for expenditure in the economy and the level of expenditures required to clear all all markets of available goods. In the long-run, changes in the quantity of money only influences the average level of prices. Finally, we have considered the role of credit markets in facilitating demand for money, as well as the role that credit markets play in generating economic fluctuations. Except in regard to the Solow model, we have ignored the role of labor in producing value in the economy. 

Wealth is generated by the contribution of value generating inputs. One such input is capital. Who ever owns capital – which includes everything from machinery to seeds for growing sunflowers to land – can either directly contribute that capital in productive processes or rent that capital to another party who will then use it in production. The value of capital is a function of its productivity while its rental price is a function of the market interest rate (as we learned last chapter in regard to wine). No matter how productive a given unit of capital is, its equilibrium rental price is a fraction of the price that capital would fetch if sold in the marketplace. 

Entrepreneurs create value by effectively aligning productive means – labor and capital – toward the provision of consumer demands. This is a risky process, which is to say that it is costly. In addition to providing for demands of consumers, entrepreneurs must also predict the demands of consumers. They arrange labor and capital in a manner that efficiently produces value.

The role of labor in the economy, then, is intimately tied to capital. Like capital, labor earns a return. That return is the wage of labor. In society where slavery does not exist, labor can only earn a wage. The wage of labor is equal to the value of its productivity over the period of consideration. In microeconomics, this is typically stated in terms of the marginal revenue product of labor. The marginal revenue product of labor is equal to the quantity of output produced by an additional unit of labor multiplied by the price of a unit of output:

>$w=MRP_L$
>
>$w=\frac{P\Delta Q}{\Delta L}$

In a world where the exercise of labor occurs jointly with capital, the marginal revenue product of labor is necessarily implicit. How much credit is owed to labor for its contribution to production? In traditional microeconomics, the identification of the marginal revenue product is simplified by presentation of a marginal product curve. As labor experiences diminishing marginal returns, it will only be employed until its wage is equal to or exceeds the value of production generated by labor.

While a firm may have an idea of how much an additional unit of labor generates in value, this may be difficult to surmise with accuracy. Firms will attempt to estimate the value of labor, however this outcome is most effectively promoted by competition within the market. In a perfectly competitive market, labor cannot be paid any more or any less than this value. This wage is arrived at by the market process. If an employee produces more value than he receives in terms of his or her current wage, that person can search for another job that offers a wage equal to his or her marginal revenue product. Likewise, other employers have incentive to offer that employee a higher wage. This also places pressure on the firm that currently employs the worker to increase his or her wages.

## Economic Growth and Wages

The wage of labor may rise due to short-run or long-run factors. If capital becomes more productive due to improvement in technology, then the wage of labor that operates that capital can be expected to increase. Thus, a person who operates heavy machinery to dig a ditch can expect to receive a wage higher than a person who uses a shovel to dig a ditch. An accountant who uses software to manage accounts can be expected to earn a higher wage than someone who writes all values and performs all calculations by hand. Expectations of high returns incentivize firms to hire greater quantities of labor and to pay that labor a higher real wage.

## The Labor Force: Its employment and composition

Economists and economic commentators often look to the values of aggregate measures of labor to infer the health of the economy. This is by no means a simply process. Economists must be careful in considering the meaning of an aggregate before interpreting changes in the value of the aggregate.

Often economists refer to the unemployment rate. This is itself an aggregate measure comprised of other aggregate measures. The unemployment rate is defined:

>$u = 1 - \frac{E}{L}$
>
>$E:$ $Number$ $Employed$
>
>$L:$ $Number$ $in$ $Labor$ $Force$

The unemployment rate is the fraction of the labor force that is not employed. This can be rewritten for clarity:

>$u = \frac{L - E}{L}$
>
>$y = \frac{U}{L}$

The unemployment rate can change either due to a change in the number employed or due to a change in labor force participation. If more individuals enter the labor force and the number of jobs available remains constant, the unemployment rate is driven up as both U and L increase by the value of the inflow of labor. Likewise, an exit of individuals from the pool of labor seeking employment promotes a fall in the the unemployment rate.

This dynamic presentation of the unemployment rate explains some fluctuations that we have observed in the unemployment rate over the last several decades. Starting in the late 1960s, the labor force participation rate began to increase. This increase did not slow until the early 1980s. From 1970 until the mid 1980s, the rate of unemployment also showed a relative increase. This can be explained largely by the relative increase in participation of female labor. Participation of women in the labor force began increasing in the late 1950s, beginning a dramatic climb in the late 1960s. In the early 1970s, many young men began to return home from Vietnam, which also put strain on the labor market. The general increase in labor required a restructuring of the economy that took more than a decade. Only once the gap between the participation rate of female labor and the participation rate of labor generally stabilized did unemployment rate begin to fall back to levels preceding this period.

The fall in the unemployment rate after 2008 can also be described in terms of a changing composition of the labor force. Over the last decade, the share of the labor force that is 55 years or older has increased. The labor force participation rate of this group is 40\%. In the case of gender, the share of the potential labor force that is male or female is about constant. A demographic shift that has resulted from the aging of baby boomers, however, has increased the significance of that group to overall labor force participation. An increased weight for a group whose labor force participation rate is significantly below the average has led to an overall reduction in the labor force participation rate. Thus, part of the fall in the unemployment is a figment of measurement. 

## Kinds of Unemployment

Not all unemployment is bad in the sense that welfare is not necessarily improved by a decrease in the rate of unemployment. Giving existing capital configurations and policies, some positive level of unemployment is efficient. This level of employment is referred to as the natural rate of unemployment. It includes unemployment that arises due to the cost of moving between jobs, either due to the need to find work that is better suited to the employee or due to the searcher taking time to build the skill required to find a job of their liking.

Unemployment that occurs due to the time required to find a suitable job is frictional unemployment. An unemployed person could become employed by taking the job that he or she can most quickly acquire. An unemployed medical doctor could conceivable acquire employment pushing a lawnmower for a landscaping company. While this may allow the doctor to earn a wage, he will likely generate less value pushing lawnmowers than he would serving patients as a doctor. Assuming he himself does not value working in landscaping enough to offset the loss of value generate from his practice as a medical doctor, it is actually inefficient for him to engage in this form of employment.

Structural employment is similar to frictional unemployment. Rather than referring to unemployed labor that is between jobs due solely to a search for a new job without a change in the field, the person unemployed must acquire a new skill in order to find his or her preferred job as their old skills are no longer suitable for earning a wage at the level he or she prefers. The structure of the economy does not match his or her current skill set. Just as it is efficient for the individual experiencing frictional unemployment to remain unemployed, it is often efficient for a person who is capable of and desires to learn new skills to earn a higher wage to invest their time in building up their own abilities.

Not all forms of employment are efficient in the sense that they would persist in macroeconomic equilibrium where there exists no excess supply of or excess demand for final goods and services. Unemployment that occurs due to the existence of macroeconomic disequilibrium is known as cyclical unemployment. Cyclical unemployment can be positive or negative. ## Unions, Minimum Wage, and Labor Dislocation

One form of structural unemployment occurs occurs as a result of implementation of a minimum wage. This can occur either by legal statute or due to unionization of labor. In the first case, the minimum wage cannot usually be avoided. The result is that fewer labor can be employed. A unionized sector, such as steel in the U.S., will raise wages, constrains the quantity of labor employed in that sector (Figure 7). This leads inevitably to one of two outcomes. If all sectors of the labor market are unionized, labor disemployed due to the minimum wage will remain unemployed. If there are sectors of the labor market that are not unionized, those disemployed due to unionization may seek employment in non-unionized sectors (Figure 8). The quantity of labor employed in those sectors increases as a result, and wages in the sectors fall from their initial level.

Intervention into the labor market tends not to improve the welfare of labor overall as a class. More often, it results in a redistribution of wealth and a loss of wealth for both firm’s and labor overall. Those who benefit from this redistribution have incentive to support it. In the case of a legislated minimum wage, support may come from those firms who can afford to pay a high minimum wage in hope that smaller competitors will suffer a greater burden from the policy. Thus, support for the minimum wage from leaders of large corporations is of no surprise. Whether or not they realize, leaders of these organizations are incentivized to support such policies.

## Inflation and Unemployment: Phillips Curve

That there exists a relationship between short-run macroeconomic fluctuations and the level of employment has led some policymakers and theorists to focus attention on the effects of monetary policy on unemployment. While part of the fluctuations in unemployment are simply due to changes in the marginal productivity of capital, much as was recognized by John Maynard Keynes in his General Theory, part of this fluctuation in unemployment is also due to a systematic fooling of expectations of labor. The Phillips Curve identifies the effect on unemployment of a discrepancy between expected inflation and observed inflation. If observed inflation is higher than the level of inflation expected by labor, then the real wages of labor will fall. This represents a movement along the demand curve for labor such that firms are able to hire more labor than is sustainable in equilibrium. 

The Phillips Curve represents the supply of labor given an average level of expected inflation for labor. As actual inflation rises above expected inflation, the rate of unemployment falls. This simply means that the rate of employment is rising. If the horizontal axis represented employment, the short-run Phillips Curve would represent an increase in the quantity of labor supplied as observed inflation rises above expected inflation. Notice that this is different the the short-run excess demand that may arise with inflation. While an excess demand for goods may have a positive effect on employment, the Phillips Curve only considers unemployment changes that result from the discrepancy between observed inflation and the inflation rate expected by labor. Thus, when the effect from the short-run Phillips Curve is operating even when the price level approaches a value consistent with monetary equilibrium. Only after inflation expectations of labor adjust does the short-run Phillips Curve shift, moving unemployment back the the natural rate represented by the long-run Phillips Curve.

Just as the rise of inflation above expectations can promote a fall in unemployment, a fall in inflation below expectations or a rise in expected inflation above observed inflation will lead to an increase in the rate of unemployment. Only once inflation expectations adjust to match observed conditions can the rate of unemployment fall to its natural rate.
The short-run relationship between inflation and unemployment breaks down due to adjustment of expectations. A comparison between these rates from 1948 to present show this breakdown. While a noticeable relationship may exist for as long as a few years, the collection of dots appear more as a blob than as any identifiable pattern. 

Compare this to the relationship between real income and unemployment, which appear to be negative. The highest levels of unemployment occur during periods of contraction of real income whereas the lowest levels of unemployment appear in years with positive real GDP growth. The effects observed by the Phillips Curve are emprically transient, whereas the long-run relationship between real income growth and unemployment appear more consistently. Observations such as this have served as foundation for a critique of activist monetary policy in favor of policies that lower the costs of doing business that are incurred, for example,  due to an inefficient regulatory environment.

## Conclusion

The wage of labor and the quantity of labor employed can also be influenced by short-run factors. During a business cycle driven by fluctuations in aggregate demand, the value of the marginal product of capital tends to fluctuate, leading demand for labor by firms to fluctuate as well. Employment of labor and its wage may also fluctuate due to discrepancies between observed and expected inflation. When inflation rises above expected inflation, this tends to lead to higher levels of employment. If inflation is below expectations, the level of employment will be negatively affected.

While employment of a given unit of labor is a microeconomic phenomenon, the market as a whole exhibits macroeconomic fluctuations as demand for labor is tied to the consumer demand for goods and services.
